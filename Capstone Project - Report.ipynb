### Methodology

I first used BeautifulSoup package to scarpe the list of top 50 college towns fromn http://www.collegeranker.com/features/best-college-towns/ . I then used Google's Geocoder to get the longitude and latitude of all college towns so that I could visualize it on a map. After that I used FourSquare's explore API to get the data of the top 100 trending venues of each town. Then I used sci-kit learns One-Hot encoding to turn the categorical data into numerical data so that I could use the K-means clustering technique to group similar towns.I devided to use the Kmeans clustering technique as this is an unsupervised problem and would be best fitted to compare towns using venue data. I then chose the cluster that contained Flagstaff, AZ. Since we were looking to open a coffee shop, I chose the town with no coffee shops listed in the Top 10 most common venues, Asheville, NC.In order to figure what section of Asheville, NC that the coffee shop should opened in I used a package called uszipcode to oull all the zip codes in Asheville. I then used FourSquare data to pull all venues by search term "coffee" for each zip code. I used Follium to plot the location of each venue by zipcode.
